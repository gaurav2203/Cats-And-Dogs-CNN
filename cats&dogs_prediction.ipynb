```python
from keras.models import load_model
import cv2
import matplotlib.pyplot as plt
```

```python
model= load_model('cats&dogs.h5')
```

```python
classes= {
    0: 'Dog',
    1: 'Cat'
}
```

```python
def Final(image_path):
    image= cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    img= cv2.imread(image_path)
    image= cv2.resize(image, (100,100)).reshape(1, 100, 100, 1)
    prediction= model.predict(image)
    rounded= round(prediction[0][0])
    print(classes[rounded])
    plt.imshow(img)
Final('./train/dog.2.jpg')
```
